```python
url = "https://uk.indeed.com/jobs?q=web+automation&l=United+Kingdom&sc=0kf%3Aattr%28DSQF7%29%3B&start="
all_urls = []

for page in range(10,100,10):
    next_urls = url + str(page) + "&pp=gQAPAAAAAAAAAAAAAAAB9r181AAeAQALtt_sJdfhSC-54yia6ATKkYvA0BE5Hz30MsXcAAA&vjk=42eba9e695c59f53"
    all_urls.append(next_urls)

for url in all_urls:
    # render = requests.get(url)
    # the_html = beauty(render.content, "html.parser")
    
    pag = requests.get(url)
    soup = BeautifulSoup(pag.content, 'html.parser')
    links = []
    for link in soup.find_all('a'):
        links.append(link.get('href')) #finds all link with attr href
        # print(links)
        reds= links
    with open('indeed.txt', 'a') as f:
            #   f.write(' ')  
        f.write(str(reds))
```
